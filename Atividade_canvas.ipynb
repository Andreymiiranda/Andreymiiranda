# Correções

```python
#Tabela customer:  Valor ausente na Coluna "age":
media_idade = df_customer['age'].mean()
df_customer["age"] = df_customer['age'].fillna(media_idade)
display(df_customer)
```

```python
#Tabela rentals: daily_rate <0
df_rental.loc[df_rental["daily_rate"]<=0,"daily_rate"] = pd.NA
display(df_rental)
```

```python
#Tabela vehicles - c02_g_per_km dado ausente
df_vehicles['co2_g_per_km'] = df_vehicles['co2_g_per_km'].fillna(120)
display(df_vehicles)
# ou tirando a média:
#media_co2 = df_vehicles['co2_g_per_km'].mean()
#df_vehicles['co2_g_per_km'] = df_vehicles['co2_g_per_km'].fillna(media_co2)
```

```python
#Tabela rentals  - daily_rate está como string:
df_rental['daily_rate'] = pd.to_numeric(df_rental["daily_rate"],errors="coerce")
display(df_rental)
```

```python
#Tabela rentals - data inválida:
df_rental["start_date"] = pd.to_datetime(df_rental["start_date"],errors="coerce")
df_rental["end_date"] = pd.to_datetime(df_rental["end_date"],errors="coerce")
display(df_rental)
```

Engenharia de dados


*   Calcule a duração de cada locação(em dias)
*   Calcule a receita total por locação.
*   Faça Merge com a base veículos.
*   Calcule a emissão de CO2 por locação.
*   Extraia o mês e data de início de locação.



```python
#Criando Backup de dados
df_carbon_clean = df_carbon
df_rental_clean = df_rental
df_customer_clean = df_customer
df_vehicles_clean = df_vehicles
```

```python
#Calculando a duração de cada locação(em dias)
df_rental_clean["dias_locação"] = (
    df_rental_clean["end_date"] - df_rental_clean["start_date"]).dt.days
display(df_rental_clean)
```

```python
#Calculando a receita total por locação.
df_rental_clean["total_receita"] = df_rental_clean["dias_locação"] * df_rental_clean['daily_rate']
display(df_rental_clean)
```

```python
#Faça Merge com a base veículos.
df_consolidado = df_rental_clean.merge(df_vehicles_clean, on="vehicle_id")
df_consolidado = df_consolidado.merge(df_customer_clean, on="customer_id")
```

```python
#Calcule a emissão de CO2 por locação.
df_consolidado["co2 emitido"] = df_consolidado["co2_g_per_km"] * df_consolidado["distance_km"]
display(df_consolidado)
```

```python
# Extraia ano e mês da data de início da locação.
# para extrair informções da data temos
# .dt.days - dia
# .dt.month - mes
# .dt.year - ano
# .dt.weekday - dia da semana
df_consolidado['mes'] = df_consolidado['start_date'].dt.month
df_consolidado['ano'] = df_consolidado['start_date'].dt.year
df_consolidado['dia_da_semana'] = df_consolidado['start_date'].dt.weekday

display(df_consolidado)
```
